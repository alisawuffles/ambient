```python
def textify_labels(gold_string):
    gold_string = gold_string.replace('.0', '')
    for label_id, label in id2label.items():
        gold_string = gold_string.replace(str(label_id), label)
    return gold_string

def textify_disambiguation(answer_string):
    assert answer_string in ['1', '2']
    return True if answer_string == '1' else False
```

```python
gold_columns = [f'Answer.q{i}_gold' for i in range(4)]
for gold_col in gold_columns:
    batch_df[gold_col] = batch_df[gold_col].astype(str)
    batch_df[gold_col] = batch_df[gold_col].apply(textify_labels)

disambiguation_columns = [f'Answer.d{i}_gold' for i in range(1,4)]
for disam_col in disambiguation_columns:
    batch_df[disam_col] = batch_df[disam_col].astype(str)
    batch_df[disam_col] = batch_df[disam_col].apply(textify_disambiguation)
```

```python
# 'Input.id',
keep_columns = (['WorkerId', 'SubmitTime', 'Answer.ee', 'Input.id', 'Input.premise', 'Input.hypothesis', 
                 'Input.ambiguous_sent', 'Input.distractor_idx', 'Input.ambiguous_sent_html']
                + batch_df.columns[batch_df.columns.str.contains('Answer.')].tolist()
                + batch_df.columns[batch_df.columns.str.contains('interpretation')].tolist())
batch_df = batch_df[keep_columns]
column_rename_map = {c: c.split('.')[-1] for c in batch_df.columns if (c.startswith('Answer.') or c.startswith('Input.'))}
column_rename_map.update({
    'WorkerId': 'worker_id', 
    'SubmitTime': 'submit_time',
    'Answer.ee': 'time_on_page'})
batch_df.rename(columns=column_rename_map, inplace=True)

batch_df['ambiguous_sent'] = ['premise' if 'premise' in x else 'hypothesis' for x in batch_df['ambiguous_sent_html']]
batch_df.drop('ambiguous_sent_html', axis=1, inplace=True)
```

```python
gold_dict = {92281: [1,2], 76612: [1,2], 84927: [1,2], 6788: [0,1], 39483: [1,2]}
```

```python
## calculate results
num_examples = batch_df.id.nunique()
num_annotators = len(eligible_workers)

# annotations
ambiguous_X = np.empty((num_examples, num_annotators))
disambiguated_X = np.empty((num_examples*2, num_annotators))

# answer key
gold_X = np.empty((num_examples*2))

# acceptability judgments
disambiguation_Y = np.empty((num_examples*2, num_annotators))
distractor_Y = np.empty((num_examples, num_annotators))

for i, (id, example_df) in enumerate(batch_df.groupby('id')):
    distractor_idx = example_df.iloc[0].distractor_idx
    
    gold_X[i*2] = gold_dict[id][0]
    gold_X[i*2+1] = gold_dict[id][1]
    
    for j, worker_id in enumerate(eligible_workers):
        worker_row = example_df.loc[example_df['worker_id'] == worker_id].iloc[0]
        
        # original annotation
        ambiguous_X[i, j] = label2id[worker_row.q0_gold]
        
        # annotation for disambiguations
        disambiguated_ex_labels = [label2id[worker_row[f'q{i}_gold']] for i in range(1,4)]
        del disambiguated_ex_labels[distractor_idx]
        disambiguated_X[i*2, j] = disambiguated_ex_labels[0]
        disambiguated_X[i*2+1, j] = disambiguated_ex_labels[1]
        
        # distractor acceptability
        acceptability_labels = [worker_row[f'd{i}_gold'] for i in range(1,4)]
        distractor_Y[i, j] = acceptability_labels[distractor_idx]
        
        # disambiguation acceptability
        del acceptability_labels[distractor_idx]
        disambiguation_Y[i*2, j] = acceptability_labels[0]
        disambiguation_Y[i*2+1, j] = acceptability_labels[1]
```

```python
scores = []
for i in range(num_examples):
    s = 0
    for j in range(num_annotators):
        if disambiguation_Y[i*2,j] == 1.0 and disambiguation_Y[i*2+1,j] == 1.0 and disambiguated_X[i*2,j] == gold_X[i*2] and disambiguated_X[i*2+1,j] == gold_X[i*2+1]:
            s += 1
    scores.append(s)

scores
```

### grade quals HIT

```python
answer_key = {
    0: {
        'd1_gold': False,
        'd2_gold': True,
        'd3_gold': True,
        'q0_gold': ['neutral', 'entailment'],
        'q1_gold': 'neutral',
        'q2_gold': 'entailment',
        'q3_gold': 'neutral',
    },
    1: {
        'd1_gold': True,
        'd2_gold': False,
        'd3_gold': True,
        'q0_gold': ['neutral', 'entailment'],
        'q1_gold': 'entailment',
        'q2_gold': 'neutral',
        'q3_gold': 'neutral',
    },
    2: {
        'd1_gold': True,
        'd2_gold': True,
        'd3_gold': False,
        'q0_gold': ['entailment', 'neutral'],
        'q1_gold': 'entailment',
        'q2_gold': 'neutral',
        'q3_gold': 'neutral',
    },
    3: {
        'd1_gold': True,
        'd2_gold': False,
        'd3_gold': True,
        'q0_gold': ['entailment', 'neutral'],
        'q1_gold': 'entailment',
        'q2_gold': 'neutral',
        'q3_gold': 'neutral',
    },
    4: {
        'd1_gold': True,
        'd2_gold': False,
        'd3_gold': True,
        'q0_gold': ['entailment', 'contradiction'],
        'q1_gold': 'contradiction',
        'q2_gold': ['entailment', 'contradiction'],
        'q3_gold': 'entailment',
    }
}
```

```python
def score(worker_answers):
    points = 0 
    for example_id in worker_answers:
        for q in worker_answers[example_id]:
            if isinstance(answer_key[example_id][q], list):
                if worker_answers[example_id][q] in answer_key[example_id][q]:
                    points += 1
            else:
                if worker_answers[example_id][q] == answer_key[example_id][q]:
                    points += 1
    return points
```

```python
scores = {}
for worker_id in eligible_workers:
    worker_answers = {}
    for example_id in range(5):
        worker_answers[example_id] = {}
        example_row = batch_df.loc[(batch_df['worker_id'] == worker_id) & (batch_df['idx'] == example_id)].iloc[0]
        for key in ['d1_gold', 'd2_gold', 'd3_gold', 'q0_gold', 'q1_gold', 'q2_gold', 'q3_gold']:
            worker_answers[example_id][key] = example_row[key]
    scores[worker_id] = score(worker_answers)
```
