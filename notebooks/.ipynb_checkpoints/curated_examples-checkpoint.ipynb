```python
import pandas as pd
```

```python
df = pd.read_csv('annotation/AmbiEnt/curated_examples.csv')
df = df.where(pd.notnull(df), None)
df.head(3)
```

```python
row = df.iloc[0]
len(set([row[f'premise{k+1}'] for k in range(4) if row[f'premise{k+1}']]))
```

```python
# find incomplete examples
for i, row in df.iterrows():
    if isinstance(row['labels'], float) or ',' not in row['labels']:
        print(row['premise'])
    num_premise_disambiguations = len(set([row[f'premise{k+1}'] for k in range(4) if row[f'premise{k+1}']]))
    num_hypothesis_disambiguations = len(set([row[f'hypothesis{k+1}'] for k in range(4) if row[f'hypothesis{k+1}']]))
    disam_labels = set([row[f'label{k+1}'] for k in range(4) if isinstance(row[f'label{k+1}'], str)])
    if disam_labels != set(row['labels'].split(', ')):
        print(row['premise'])
    if num_premise_disambiguations == 1 or num_hypothesis_disambiguations == 1:
        print(row['premise'])
    if max(num_premise_disambiguations, num_hypothesis_disambiguations) < 2:
        print(row['premise'])
    if num_premise_disambiguations == 0 and num_hypothesis_disambiguations == 0:
        print(row['premise'])
    ambiguous_sentence_key = 'premise' if num_premise_disambiguations > 0 else 'hypothesis'
    for k in range(4):
        if type(row[f'{ambiguous_sentence_key}{k+1}']) != float:
            if type(row[f'label{k+1}']) == float:
                print(row['premise'])
```

```python
# format examples
examples = []
for i, row in df.iterrows():
    num_premise_disambiguations = len(set([row[f'premise{k+1}'] for k in range(4)]))
    num_hypothesis_disambiguations = len(set([row[f'hypothesis{k+1}'] for k in range(4)]))
    
    premise_ambiguous, hypothesis_ambiguous = [False]*2
    if num_premise_disambiguations >= 2:
        premise_ambiguous = True
    if num_hypothesis_disambiguations >= 2:
        hypothesis_ambiguous = True
    
    disambiguations = []
    for k in range(4):
        if row[f'label{k+1}']:
            d = {
                'premise': row[f'premise{k+1}'] or row['premise'],
                'hypothesis': row[f'hypothesis{k+1}'] or row['hypothesis'],
                'label': row[f'label{k+1}']
            }
            disambiguations.append(d)
    
    ex = {
        'premise': row['premise'],
        'hypothesis': row['hypothesis'],
        'premise_ambiguous': premise_ambiguous,
        'hypothesis_ambiguous': hypothesis_ambiguous,
        'labels': row['labels'],
        'meta': {'source': row['source']},
        'disambiguations': disambiguations
    }
    examples.append(ex)
```

```python
df = pd.DataFrame(examples)
df = df.reset_index()
df = df.rename({'index': 'id'}, axis=1)
df['id'] = [f'{id}_c' for id in df['id']]
```
