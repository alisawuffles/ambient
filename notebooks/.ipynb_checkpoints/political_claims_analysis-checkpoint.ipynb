```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from transformers import RobertaTokenizer, RobertaForSequenceClassification
from utils.utils import predict_nli
from sklearn.metrics import confusion_matrix, recall_score, precision_score, f1_score
```

```python
annotations = pd.read_csv('political-claims/dev-annotations.csv').drop('notes', axis=1)
preds = pd.read_json('political-claims/zero-shot-d02-p0.9-dev-results-200.jsonl', lines=True)[['example_id', 'label', 'claim', 'cleaned_claim', 'ambiguity_score', 'paraphrases']]
```

```python
preds = preds.merge(annotations[['claim', 'annotation']], how='inner', on='claim')
preds = preds.drop('claim', axis=1).rename({'cleaned_claim': 'claim'}, axis=1)
preds.head(3)
```

```python
ambiguous_ids = annotations.loc[annotations['annotation'] == 'a'].index
```

```python
preds.loc[~preds.index.isin(ambiguous_ids)].ambiguity_score.mean()
```

```python
preds.loc[preds.index.isin(ambiguous_ids)].ambiguity_score.mean()
```

```python
model = RobertaForSequenceClassification.from_pretrained('../wanli/models/roberta-large-wanli').to('cuda')
tokenizer = RobertaTokenizer.from_pretrained('../wanli/models/roberta-large-wanli')
```

```python
for i, row in preds.iterrows():
    paraphrases = []
    for p in row['paraphrases']:
        probs = predict_nli(row['claim'], p[0], model, tokenizer)
        pred = max(probs, key=probs.get)
        if pred == 'entailment':
            paraphrases.append(p)
    preds.at[i, 'paraphrases'] = paraphrases
    if len(paraphrases) == 0:
        preds.at[i, 'ambiguity_score'] = 0
    else:
        preds.at[i, 'ambiguity_score'] = max([p[1] for p in paraphrases])
```

```python
preds.to_dict(orient='records')
```

```python
precisions = []
recalls = []
f1s = []
thresholds = np.arange(0, 1.0, 0.1)
for t in thresholds:
    y_pred = [1 if a > t else 0 for a in preds.ambiguity_score]
    y_true = [1 if a == 'a' else 0 for a in preds.annotation]
    precisions.append(precision_score(y_true, y_pred))
    recalls.append(recall_score(y_true, y_pred))
    f1s.append(f1_score(y_true, y_pred))
```

```python
fig, ax = plt.subplots()
sns.lineplot(x=thresholds, y=precisions, label='precision')
sns.lineplot(x=thresholds, y=recalls, label='recall', color='yellow')
ax.set_xlabel('Threshold on ambiguity score')
```

```python
fig, ax = plt.subplots()
sns.lineplot(x=np.arange(0, 1.0, 0.1), y=f1s)
ax.set_xlabel('Ambiguity score threshold')
ax.set_ylabel('F1')
```
