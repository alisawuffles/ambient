```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 160
import numpy as np
import pandas as pd
from generation.gpt3_generation import request
from utils.utils import strip_punctuation_and_casing
```

```python
%cd '../memo-traps'
```

```python
model_sizes = ['ada', 'babbage', 'curie', 'davinci']
last_token_dicts = {}
prompt="""A rising tide tide lifts all"""

for model_size in model_sizes:
    response = request(
        prompt=prompt,
        model=model_size,
        echo=True,
        logprobs=100,
        max_tokens=1
    )
    
    last_token_dict = response['choices'][0]['logprobs']['top_logprobs'][-1]
    last_token_dicts[model_size] = dict(last_token_dict)
```

```python
fig, axes = plt.subplots(2,2,figsize=(10,10))
for i, model_size in enumerate(model_sizes):
    last_token_dict = last_token_dicts[model_size]
    ax = axes[i//2,i%2]
    last_token_dict = sorted(last_token_dict.items(), key=lambda item: item[1], reverse=True)
    x, y = [item[0] for item in last_token_dict], [np.exp(item[1]) for item in last_token_dict]
    num_tokens = 12
    sns.lineplot(x=x[:num_tokens], y=y[:num_tokens], ax=ax)
    ax.set_title(model_size)
    ax.set_xticklabels(x[:num_tokens], rotation=45)

# fig.suptitle(prompt, fontsize=15)
plt.tight_layout()
# fig.savefig(f'plots/{"_".join(strip_punctuation_and_casing(prompt).split())}', dpi=100)
```

```python
davinci_top1 = max(last_token_dicts['davinci'], key=last_token_dicts['davinci'].get)
# davinci_top1 = ' terrible'
y = []
xticklabels = model_sizes.copy()
for model in model_sizes:
    try:
        y.append(last_token_dicts[model][davinci_top1])
    except KeyError:
        xticklabels.remove(model)
        continue

fig, ax = plt.subplots()
ax.set_title(davinci_top1)
sns.lineplot(range(1, len(y)+1), y)
ax.set_ylabel('logprob')
ax.set_xticks(range(1, len(y)+1))
ax.set_xticklabels(xticklabels, rotation=30)
```

```python
model_sizes = ['ada', 'babbage', 'curie', 'davinci']
token_logprobs = {}

for model_size in model_sizes:
    response = request(
        prompt=prompt,
        model=model_size,
        echo=True,
        logprobs=1,
        max_tokens=0
    )
    tokens = response['choices'][0]['logprobs']['tokens']
    token_logprobs[model_size] = response['choices'][0]['logprobs']['token_logprobs']
```

```python
plt.figure()
fig, ax = plt.subplots(figsize=(20,5))
for i, model_size in enumerate(model_sizes):
    y = [np.exp(x) for x in token_logprobs[model_size][1:]]
    sns.lineplot(x=range(1, len(y)+1), y=y, label=model_size, ax=ax)

ax.set_xticks(range(1, len(y)+1))
ax.set_xticklabels(tokens[1:], rotation=30)
ax.legend()
ax.set_xlim(left=1)
plt.show()
```
