```python
def paraphrase_gpt3(sentence, n, top_p=0.9, engine='text-davinci-003'):
    generations = set()
    for i in range(n):
        instruction = 'Paraphrase the text.'
        context = f'{instruction}\n\n{sentence}\nParaphrase: '
        generation = request(
            prompt=context,
            model=engine,
            top_p=top_p,
            n=1,
            stop=None,
            return_only_text=True
        )
        
        generation = generation.replace('Paraphrase:', '').strip()
        generations.add(generation)

    generations.discard('')
    generations.discard(sentence)
    return generations
```

```python
gold_paraphrases = pd.read_csv('political-claims/train-paraphrases.csv')
# gold_paraphrases = pd.read_csv('analysis/paraphrases.csv')

def paraphrase_gpt3(sentence, n, num_incontext_examples=5, paraphrase_type='generic'):
    generations = set()
    for i in range(n):
        if paraphrase_type == 'generic':
            instruction = 'Paraphrase the text.'
            in_context_examples = gold_paraphrases.sample(frac=1).drop_duplicates(subset='text').sample(5)
        elif paraphrase_type == 'literal':
            instruction = 'Write a very literal paraphrase of the text.'
            in_context_examples = gold_paraphrases.loc[gold_paraphrases['type'] == 'literal'].sample(5)
        elif paraphrase_type == 'pragmatic':
            instruction = 'Write a likely paraphrase of the text.'
            in_context_examples = gold_paraphrases.loc[gold_paraphrases['type'] == 'pragmatic'].sample(5)
        in_context_examples = [(ex['text'], ex['paraphrase']) for ex in in_context_examples.to_dict(orient='records')]
        context = f"{instruction}"
        for j, e in enumerate(in_context_examples):
            context += f'\n{j+1}. {e[0]}\nParaphrase: {e[1]}'
        context += f'\n{num_incontext_examples+1}. {sentence}'
        generation = request(
            prompt=context,
            model='text-davinci-003',
            top_p=0.95,
            n=1,
            stop=None,
            return_only_text=True
        )
        
        generation = generation.replace('Paraphrase:', '').strip()
        generations.add(generation)

    generations.discard('')
    generations.discard(sentence)
    return generations
```

```python
claim = "We have a system now where in 40 states, the highest-paid public employee is the state university’s head football or basketball coach."
paraphrase_gpt3(claim, n=5)
```

```python
multilabel_model = RobertaForSequenceClassification.from_pretrained('models/roberta-large-wanli-multilabel').to('cuda')
tokenizer = RobertaTokenizer.from_pretrained('models/roberta-large-wanli-multilabel')
```

```python
def predict_nli(premise, hypothesis, model, tokenizer):
    x = tokenizer(premise, hypothesis, return_tensors='pt', max_length=128, truncation=True).to('cuda')
    logits = model(**x).logits
    # multi-task model
    if hasattr(model, 'output_heads'):
        probs = logits.softmax(dim=-1).squeeze(0)
        return {model.config.id2label[i]: probs[i,1].item() for i in range(len(probs))}
    # multi-label model
    elif model.config.problem_type == 'multi_label_classification':
        logits = logits.squeeze(0)
        probs = sigmoid(logits)
        return {model.config.id2label[i]: probs[i].item() for i in range(len(probs))}
    # classification model
    else:
        probs = logits.softmax(dim=1).squeeze(0)
        return {model.config.id2label[i]: probs[i].item() for i in range(len(probs))}
```

```python
sns.set_style('white')
sns.set_palette('muted')
red = sns.color_palette('muted')[3]
green = sns.color_palette('muted')[2]
gray = sns.color_palette('muted')[-3]
```

```python
premise = 'When San Francisco banned plastic grocery bags, you saw the number of instances of people going to the ER with things like salmonella and other related illnesses spike.'
hypothesis = """When San Francisco banned plastic grocery bags, it caused the number of instances of people going to the ER with things like salmonella and other related illnesses spike."""
probs = predict_nli(premise, hypothesis, multilabel_model, tokenizer)
print(probs)
x = list(probs.values())
fig, ax = plt.subplots(figsize=(4,2))
sns.barplot(x, NLI_LABELS, order=['entailment', 'neutral', 'contradiction'], palette=[green, gray, red])
ax.set_xlim(right=1)
```

## detect misleading political claims

```python
claims_df = pd.read_json('data/claim-decomp/dev.jsonl', lines=True)
# claims_df.columns = ['claimID', 'claim', 'claimURL', 'reason', 'categories', 'speaker', 'checker', 'tags', 'articleTitle', 'publishDate', 'claimDate', 'entities']
# claims_df = claims_df.loc[claims_df['claimID'].str.startswith('pomt')]
```

```python
strings_to_remove = ['Says that', 'Says', '“', '"', '”', '(', ')']

claims_df['cleaned_claim'] = None
claims_df['ambiguity_score'] = None
claims_df['paraphrases'] = None
claims_df['paraphrases'] = claims_df['paraphrases'].astype('object')
for i, row in tqdm(claims_df.iterrows(), total=len(claims_df.index)):
    claim = row['claim']
    for s in strings_to_remove:
        claim = claim.replace(s, '')
    claim = claim.strip()
    claim = claim[0].upper() + claim[1:]
    paraphrases = paraphrase_gpt3(claim, n=5, engine='text-davinci-003', top_p=0.4)
    scores = []
    for paraphrase in paraphrases:
        probs = predict_nli(claim, paraphrase, multilabel_model, tokenizer)
        s = sorted([p for p in probs.values()], reverse=True)[1]
        scores.append(s)
    claims_df.at[i, 'cleaned_claim'] = claim
    claims_df.at[i, 'paraphrases'] = [(p,s) for p,s in zip(paraphrases, scores)]
    claims_df.at[i, 'ambiguity_score'] = np.nanmax(scores)
```

```python
claims_df.to_json('political-claims/zero-shot-d03-p0.4-dev-results-200.jsonl', lines=True, orient='records')
```
