```python
import pandas as pd
from evaluation.generative_evaluation import generation_to_lines, label_verbalizer
from evaluation.edit_f1 import get_edit_f1
import numpy as np
```

```python
df = pd.read_json('results/generative_evaluation/llama-65b-n4.jsonl', lines=True)
df.head(3)
```

```python
new_results = []
for i, row in df.iterrows():
    ambiguous_sentence_key = 'premise' if row['premise_ambiguous'] else 'hypothesis'    
    generation = row['generation'].strip().split('\n\n')[0]
    generated_lines = generation_to_lines(generation)
    
    pred_rewrites = {}
    for line in generated_lines:
        for label, verbalizer in label_verbalizer.items():
            if verbalizer in line:
                pred_rewrites[label] = line.replace(verbalizer, '').strip()
    
    gold_rewrites = row['gold_rewrites']
    
    edit_f1s = []
    for label in gold_rewrites:
        if label in pred_rewrites:
            edit_f1s.append(get_edit_f1(row[ambiguous_sentence_key], gold_rewrites[label], pred_rewrites[label]))
        else:
            edit_f1s.append(0)
    
    row = row.to_dict()
    row['predicted_rewrites'] = pred_rewrites
    row['edit_f1'] = np.mean(edit_f1s)
    new_results.append(row)
```

```python
new_results.to_json('generative_evaluation/llama-65b-n4.jsonl', lines=True, orient='records')
```
