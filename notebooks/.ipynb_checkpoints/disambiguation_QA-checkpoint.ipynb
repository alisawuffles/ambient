```python
import pandas as pd
import json
```

```python
with open('data/disambiguation_qa/raw/task.json', 'r') as fin:
    data = fin.read()
```

```python
data = json.loads(data)
examples = data['examples']
```

```python
nli_examples = []
for row in examples:
    if row['target_scores']['Ambiguous'] == 1:
        for targets in row['target_scores'].items():
            if targets[0] == 'Ambiguous':
                continue
            ex = {
                'premise': row['input'],
                'hypothesis': targets[0],
                'label': 'entailment, contradiction'
            }
            nli_examples.append(ex)
```

```python
nli_examples.to_csv('data/disambiguation_qa/ambiguous_examples.csv', index=False)
```
