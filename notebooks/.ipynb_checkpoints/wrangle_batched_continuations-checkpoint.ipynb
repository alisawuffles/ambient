```python
import pandas as pd
import os
from evaluation.utils import create_test_instances
import numpy as np
```

```python
test_df = pd.read_json('annotation/AmbiEnt/test.jsonl', lines=True)
test_df = test_df[test_df['premise_ambiguous'] | test_df['hypothesis_ambiguous']]
test_instances = create_test_instances(test_df)
```

```python
len(test_instances)
```

```python
model_dir = Path('results/continuation_evaluation/flan-t5-xxl')
results_files = [d for d in os.listdir(model_dir) if 'results_' in d]
dfs = []

for file in results_files:
    df = pd.read_json(model_dir / file, lines=True)
    dfs.append(df)
```

```python
result_df = pd.concat(dfs)
len(result_df)
```

```python
result_df = result_df.loc[~result_df.duplicated(subset='ambiguous_sentence')]
len(result_df)
```

```python
total_ids = set([e['id'] for e in test_instances])
done_ids = set(result_df.id.tolist())
assert done_ids.issubset(total_ids)
missing_ids = total_ids - done_ids
```

```python
len(missing_ids)
```

```python
result_df.to_json(model_dir / 'results.jsonl', lines=True, orient='records')
```

```python
missing_df = test_df.loc[test_df['id'].isin(missing_ids)]

batch_size = 10
num_examples = len(missing_df.index)

for j, i in enumerate(np.arange(0, num_examples, batch_size)):
    missing_df.iloc[i:np.min([i+batch_size, num_examples])].to_json(f'annotation/AmbiEnt/llama_missing_shards/test_{16+j+1}.jsonl', lines=True, orient='records')
```

```python
result_df.to_dict(orient='records')
```

```python
results_dir = Path('results/continuation_evaluation/llamas')
llama_dirs = [d for d in os.listdir(results_dir) if 'llama-65b' in d]
dfs = []
for llama_dir in llama_dirs:
    results_files = [d for d in os.listdir(results_dir/llama_dir) if 'results_' in d]
    
    for file in results_files:
        df = pd.read_json(results_dir / llama_dir / file, lines=True)
        dfs.append(df)
```

```python
result_df = pd.concat(dfs)
len(result_df)
```

```python
result_df = result_df.loc[~result_df.duplicated(subset='ambiguous_sentence')]
len(result_df)
```

```python
total_ids = set([e['id'] for e in test_instances])
done_ids = set(result_df.id.tolist())
assert done_ids.issubset(total_ids)
missing_ids = total_ids - done_ids
```

```python
len(missing_ids)
```

```python
result_df.to_json('results/continuation_evaluation/llama-65b/results.jsonl', lines=True, orient='records')
```

```python
missing_df = test_df.loc[test_df['id'].isin(missing_ids)]

batch_size = 10
num_examples = len(missing_df.index)

for j, i in enumerate(np.arange(0, num_examples, batch_size)):
    missing_df.iloc[i:np.min([i+batch_size, num_examples])].to_json(f'annotation/AmbiEnt/llama_missing_shards/test_{j+1}.jsonl', lines=True, orient='records')
```
