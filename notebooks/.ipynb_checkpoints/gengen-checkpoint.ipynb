```python
import pandas as pd
import numpy as np
from transformers import RobertaTokenizer, RobertaForSequenceClassification
from utils.utils import predict_nli
from collections import Counter
from tqdm import tqdm
```

```python
df = pd.read_csv('data/gengen/GenGenpredictions-predictions_withTypeInfo.csv')
df = df.loc[df['template_number'].isin([3,4])]
```

```python
wanli_model = RobertaForSequenceClassification.from_pretrained('alisawuffles/roberta-large-wanli').to('cuda')
wanli_tokenizer = RobertaTokenizer.from_pretrained('alisawuffles/roberta-large-wanli')

mnli_model = RobertaForSequenceClassification.from_pretrained('roberta-large-mnli').to('cuda')
mnli_tokenizer = RobertaTokenizer.from_pretrained('roberta-large-mnli')
```

```python
key = 'wanli_prediction'
model, tokenizer = wanli_model, wanli_tokenizer

instantiations_df[key] = None
for i, row in tqdm(df.iterrows(), total=len(instantiations_df.index)):
    probs = predict_nli(row['generic'], row['instantiation'], model, tokenizer)
    pred = max(probs, key=probs.get)
    instantiations_df.at[i, key] = pred
```

```python
key = 'wanli_prediction'
model, tokenizer = wanli_model, wanli_tokenizer

df[key] = None
for i, row in tqdm(df.iterrows(), total=len(df.index)):
    type_sentence = row['type_sentence'] if row['type_sentence'][0] == ' ' else ' ' + row['type_sentence']
    # without type statement
    probs = predict_nli(row['generic'], row['exception'], model, tokenizer)
    pred = max(probs, key=probs.get)
    df.at[i, 'original_' + key] = pred
    # with type statement
    probs = predict_nli(row['generic'] + type_sentence, row['exception'], model, tokenizer)
    pred = max(probs, key=probs.get)
    df.at[i, key] = pred
```

```python
label_map = {'CONTRADICTION': 'contradiction', 'ENTAILMENT': 'entailment', 'NEUTRAL': 'neutral'}
exceptions_df = df.replace({
    'wanli_prediction' : label_map, 
    'mnli_prediction': label_map,
    'original_wanli_prediction': label_map,
    'original_mnli_prediction': label_map,
})
exceptions_df
```

```python
{k: np.round(v/len(exceptions_df.index), 3) for k, v in exceptions_df.original_wanli_prediction.value_counts().items()}
```

```python
exceptions_df['original_wanli_prediction'] = exceptions_df['wanli_prediction']
exceptions_df['original_mnli_prediction'] = exceptions_df['mnli_prediction']
```
