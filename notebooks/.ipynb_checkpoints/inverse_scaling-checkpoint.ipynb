## scrape web for long-reigning leaders and monarchs

```python
scraped_names = {}
```

```python
year_cutoff = 2000
```

```python
## monarchs
page = requests.get(f'https://en.wikipedia.org/wiki/List_of_current_reigning_monarchs_by_length_of_reign')
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
tables = soup.find_all('table', class_='wikitable sortable plainrowheaders')

monarch_data = []
for table in tables:
    table_body = table.find('tbody')
    rows = table_body.find_all('tr')
    rows = rows[2:] # skip header rows

    for row in rows:
        cols = row.find_all('td')
        if len(cols) != 6:
            continue
        rank, state, born, current_age, acceded, years_reigned = cols
        name = row.find('th').text.strip()
        cols = [acceded.text.strip(), name, state.text.strip()]
        
        year = int(re.sub('\[.*\]', '', cols[0].split()[-1]))
        if year > year_cutoff:
            continue
        monarch_data.append(cols)
```

```python
## state leaders
page = requests.get(f'https://en.wikipedia.org/wiki/List_of_current_state_leaders_by_date_of_assumption_of_office')
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
tables = soup.find_all('table', class_='wikitable sortable')

leader_data = []
for table in tables:
    table_body = table.find('tbody')
    rows = table_body.find_all('tr')
    rows = rows[1:] # skip header rows

    for row in tqdm(rows):
        cols = row.find_all('td')
        if len(cols) != 4:
            continue
        assumed_office, leader, state, office = cols
        cols = [re.sub('\[.*\]', '', ele.text.strip()) for ele in [assumed_office, leader, state]]
        # if holding multiple offices, get longest-held
        if type(office) != str:
            office = office.get_text(strip=True, separator='\n').splitlines()[0]
            
        if len(cols[0].split()) == 2:
            continue
        else:
            year = int(re.sub('\[.*\]', '', cols[0].split()[-1]))
            if year > year_cutoff:
                continue
        leader_data.append(cols + [office])
```

```python
formatted_examples = []
for row in leader_data:
    leader = row[1]
    country = row[2]
    position = row[-1].replace('Acting ', '').replace('-regent', '')
#     ex1 = f'{leader} was succeeded by [NAME], who is the current {position} of {country}.\nQuestion: Who is the current {position} of {country}?\nAnswer:'
#     ex2 = f'[NAME] is the current {position} of {country}, after succeeding {leader}.\nQuestion: Who is the current {position} of {country}?\nAnswer:'
    ex1 = f'{leader} was succeeded by [NAME], who is the current {position} of {country}. Which statement is correct?\n\n'
    ex2 = f'[NAME] is the current {position} of {country}, after succeeding {leader}. Which statement is correct?\n\n'
    for ex in [ex1, ex2]:
        formatted_examples.append({
            'prompt': ex,
            'original_ending': ' ' + leader,
            'new_ending': ' [NAME]',
            'country': country,
            'position': position,
        })

for row in monarch_data:
    monarch = row[1]
    country = row[2]
    ex1 = f'{monarch} was succeeded by [NAME], who is the current monarch of {country}. Which statement is correct?\n\n'
    ex1 = f'[NAME] is the current monarch of {country}, after succeeding {monarch}. Which statement is correct?\n\n'
    for ex in [ex1, ex2]:
        formatted_examples.append({
            'prompt': ex,
            'original_ending': ' ' + monarch,
            'new_ending': ' [NAME]',
            'country': country,
            'position': 'monarch'
        })
```

```python
def get_names_for_country(country, gender=None):
    """
    gender = F or M or None
    """
    def get_gender(name):
        return 'F' if 'nina' in name.find('img')['src'] else 'M'
    
    names_dict = {}
    country = country.lower()
    for webpage in ['surnam', 'firstnam']:
        try:
            page = requests.get(f'https://{webpage}.es/{"-".join(country.split())}')
        except ConnectionError:
            print(f'No page for {country}')
            return
        soup = BeautifulSoup(page.content, 'html.parser')
        names = soup.find_all('li', class_='list-item col-lg-6 mb-1')
        cleaned_names = []
        for name in names:
            # if names of a particular gender are wanted, then only include those
            if webpage == 'firstnam' and gender:
                name_gender = get_gender(name)
                if name_gender != gender:
                    continue
            cleaned_names.append(name.text.strip().split(' (')[0])
            
        names_dict[webpage] = cleaned_names
        if len(cleaned_names) < 10:
#             print(f'Not enough names for {country}')
            return
    return names_dict
```

```python
num_fake_names = 5
formatted_examples = []
for i, row in tqdm(df.iterrows(), total=len(df.index)):
    leader, position, country = row['original_ending'].strip(), row['position'], row['country']
    if position == 'Queen':
        gender = 'F'
    elif position in ['King', 'Prince', 'Sultan']:
        gender = 'M'
    else:
        gender = None
    
    # check if the country + gender pair has already been scraped, else scrape
    if (row['country'], gender) in scraped_names:
        names_dict = scraped_names[(row['country'], gender)]
    else:
        names_dict = get_names_for_country(row['country'], gender=gender)
        scraped_names[(row['country'], gender)] = names_dict
        
    if names_dict is None:
        continue
    fake_names = zip(random.sample(names_dict['firstnam'], num_fake_names), random.sample(names_dict['surnam'], num_fake_names))
    for fake_name in fake_names:
        name_string = ' '.join(fake_name)
        correct_option = f'[NAME] is the {position} of {country}.'
        wrong_option = f'{leader} is the {position} of {country}.'
        correct_answer = row['new_ending'].replace('[NAME]', name_string)
        format_option = random.choice([0,1])
        
        if format_option == 0:
            prompt = row['prompt'] + 'A. ' + correct_option +'\nB. ' + wrong_option + '\n\nAnswer:'
            prompt = prompt.replace('[NAME]', name_string)
            ex = {
                'prompt': prompt,
#                 'classes': [row['original_ending'], correct_answer],
                'classes': [' A',' B'],
                'answer_index': 0
            }
        else:
            prompt = row['prompt'] + 'A. ' + wrong_option +'\nB. ' + correct_option + '\n\nAnswer:'
            prompt = prompt.replace('[NAME]', name_string)
            ex = {
                'prompt': prompt,
#                 'classes': [correct_answer, row['original_ending']],
                'classes': [' A',' B'],
                'answer_index': 1
            }
        formatted_examples.append(ex)
```

```python
from transformers import BertForMaskedLM, BertTokenizer
```

```python
quotes = pd.read_json('raw_data/quotes.jsonl', lines=True)
```

```python
model = BertForMaskedLM.from_pretrained("bert-large-uncased", forced_bos_token_id=0)
tok = BertTokenizer.from_pretrained("bert-large-uncased")
```

```python
# use quotes that davinci prefers to ada
stems = [' '.join(quote.split(' ')[:-1]) for quote in quotes.quote]
endings = [' ' + quote.split(' ')[-1] for quote in quotes.quote]
```

```python
filtered_idxs = [i for i, (q,ce1,ce2) in enumerate(zip(quotes.quote, davinci_ce, ada_ce)) if ce1<ce2]
good_quotes = quotes.loc[filtered_idxs]
good_quotes
```

```python
examples_for_review = []
for i, row in tqdm(good_quotes.iterrows()):
    quote = row['quote'][:-1]
    assert row['quote'][-1] == '.'
    prompt, original_ending = ' '.join(quote.split(' ')[:-1]), ' ' + quote.split(' ')[-1]
    masked_quote = prompt + ' [MASK].'
    inputs = tok(masked_quote, return_tensors="pt")
    with torch.no_grad():
        logits = model(**inputs).logits

    # retrieve index of [MASK]
    mask_token_index = (inputs.input_ids == tok.mask_token_id)[0].nonzero(as_tuple=True)[0]
    
    predicted_token_ids = torch.multinomial(F.softmax(logits[0, mask_token_index], dim=-1), num_samples=5).squeeze(0)
    alternatives = [tok.decode([predicted_token_id]) for predicted_token_id in predicted_token_ids]
    alternatives = [t for t in alternatives if ' ' not in t] # ensure it's a single word
    
    # only consider alternatives that ada prefers to davinci
    davinci_ce = cross_entropy(
        [prompt] * len(alternatives),
        [' ' + a for a in alternatives], 
        engine='davinci'
    )
    ada_ce = cross_entropy(
        [prompt] * len(alternatives),
        [' ' + a for a in alternatives], 
        engine='ada'
    )
    good_alternatives = [a for a, ce1, ce2 in zip(alternatives, davinci_ce, ada_ce) if ce2 < ce1]
    
    examples_for_review.append({
        'prompt': prompt,
        'alternative_options': ', '.join(good_alternatives),
        'original_ending': original_ending
    })
```
