```python
import pandas as pd
from tqdm import tqdm
import torch
import os
from utils.constants import NLI_LABELS, id2label
from utils.utils import ensure_dir, example_revised, example_untouched
from collections import Counter
from scipy.stats import entropy
import numpy as np
import math
from transformers import RobertaTokenizer, RobertaForSequenceClassification
from modeling.multitask_model import RobertaForMultitaskSequenceClassification
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, log_loss, multilabel_confusion_matrix, classification_report
from collections import defaultdict
from utils.utils import predict_nli
from torch import sigmoid
```

```python
multitask_model = RobertaForMultitaskSequenceClassification.from_pretrained('models/roberta-large-wanli-multitask').to('cuda')
multilabel_model = RobertaForSequenceClassification.from_pretrained('models/roberta-large-wanli-multilabel').to('cuda')
```

```python
original_model = RobertaForSequenceClassification.from_pretrained('../wanli/models/roberta-large-wanli').to('cuda')
```

```python
tokenizer = RobertaTokenizer.from_pretrained('models/roberta-large-wanli')
```

```python
sns.set_style('white')
sns.set_palette('muted')
red = sns.color_palette('muted')[3]
green = sns.color_palette('muted')[2]
gray = sns.color_palette('muted')[-3]
```

```python
def predict_nli(premise, hypothesis, model, tokenizer):
    x = tokenizer(premise, hypothesis, return_tensors='pt', max_length=128, truncation=True).to('cuda')
    logits = model(**x).logits
    # multi-task model
    if hasattr(model, 'output_heads'):
        probs = logits.softmax(dim=-1).squeeze(0)
        return {model.config.id2label[i]: probs[i,1].item() for i in range(len(probs))}
    # multi-label model
    elif model.config.problem_type == 'multi_label_classification':
        logits = logits.squeeze(0)
        probs = sigmoid(logits)
        return {model.config.id2label[i]: probs[i].item() for i in range(len(probs))}
    # classification model
    else:
        probs = logits.softmax(dim=1).squeeze(0)
        return {model.config.id2label[i]: probs[i].item() for i in range(len(probs))}
```

```python
examples_df = pd.read_json('generated_data/n5_nn_davinci_noseed/filtered_examples.jsonl', lines=True)
```

```python
td_metrics = pd.read_json('../wanli/models/wanli-103079-results/roberta-large-mnli/training_dynamics_gen/td_metrics.jsonl', lines=True)
examples_df = examples_df.merge(td_metrics[['guid', 'max_variability']], left_on='id', right_on='guid', how='inner').drop('guid', axis=1)
```

```python
examples_df['ambiguity_score'] = None
examples_df['predicted_labels'] = None
for i, row in tqdm(examples_df.iterrows(), total=len(examples_df.index)):
    premise, hypothesis = row['premise'], row['hypothesis']
    probs = predict_nli(premise, hypothesis, multitask_model, tokenizer)
    preds = set([l for l, p in probs.items() if p > 0.05])
    s = np.mean(sorted([p for p in probs.values() if p > 0.05], reverse=True)[1:])
    examples_df.at[i, 'ambiguity_score'] = s
    examples_df.at[i, 'predicted_labels'] = ', '.join(sorted(preds))
```
