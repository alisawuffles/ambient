```python
import random
import pandas as pd
from tqdm import tqdm
```

```python
n_classes = 15
students_per_role = 2

students = list(range(n_students))
classes = list(range(n_classes))
roles = ['original_author_1', 'scientific_peer_reviewer', 'archaeologist', 'imaginative_researcher', 'original_author_2']
students = [
    'Orevaoghene',
    'Anas',
    'Tong',
    'Daniel',
    'Matt',
    'Victoria',
    'Shangbin',
    'Arya Krisna',
    'Minh',
    'Jaehun',
    'Moe',
    'Joongwon (Daniel)',
    'Sneha',
    'Margaret',
    'Zeyu',
    'Zucks Zixuan',
    'Shane',
    'Michael',
    'Lucille',
    'Kai',
    'Shubhkarman',
    'Taylor',
    'Sahil',
    'Xinyan (Velocity)',
    'Muru',
    '[MASK1]',
    '[MASK2]',
    '[MASK3]',
    '[MASK4]',
    '[MASK5]'
]

n_students = len(students)
n_roles = len(roles)
```

```python
assignments = {c: {f'{i}'+r: [] for r in roles for i in range(students_per_role)} for c in classes}
assignments
```

```python
students_tracker = students * int(n_classes * n_roles * students_per_role / n_students)
random.shuffle(students_tracker)
dummy_student = 'dummy'

for c in tqdm(assignments):
    class_assignments = set([dummy_student])
    for r in roles:
        # pick student1
        student1 = dummy_student
        while student1 in class_assignments:
            student1 = random.sample(students_tracker, 1)[0]
        
        class_assignments.add(student1)
        students_tracker.remove(student1)
        assignments[c]['0'+r] = student1
        
        # pick student2
        student2 = student1
        while student2 in class_assignments:
            student2 = random.sample(students_tracker, 1)[0]
        
        class_assignments.add(student2)
        students_tracker.remove(student2)
        
        assignments[c]['1'+r] = student2
```

```python
pd.DataFrame(assignments).transpose().to_csv('/gscratch/xlab/alisaliu/nlp_assignments.csv', index=False)
```
