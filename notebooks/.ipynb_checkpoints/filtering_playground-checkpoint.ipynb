```python
generations_df = pd.read_json('generated_data/wanli_disagreement_p0.9_davinci-002/examples.jsonl', lines=True)
```

```python
forbidden_stems = [
    'I think',
    'The study found that',
    'In a sense,',
    'In a way,'
]
```

```python
idxs = []
for i, row in tqdm(generations_df.iterrows()):
    premise, hypothesis = row['premise'], row['hypothesis']
    if np.any([strip_punctuation_and_casing(premise) == strip_punctuation_and_casing(b+' '+hypothesis) for b in forbidden_stems]):
        idxs.append(i)
```

```python
generations_df.loc[random.sample(idxs,1)].iloc[0].to_dict()
```

```python
generations_df.loc[generations_df['hypothesis'].str.contains("go to the party")].to_dict(orient='records')
```
