```python
import pandas as pd
from evaluation.pmi import cross_entropy
from generation.gpt3_generation import request
import random
import seaborn as sns
import matplotlib.pyplot as plt
from tqdm import tqdm
import numpy as np
import os
```

```python
ambiguous_sent = 'She must work hard.'
disambiguations = [
    'She probably works hard.',
    'It is probably the case that she works hard.',
    'She needs to work hard.',
    'It is crucial for her to work hard.',
    'It is imperative that she works hard.',
    'She has got to work hard.'
]
engine = 'davinci'
```

```python
def add_underscores_in_spaces(sentence):
    return '_'.join(sentence[:-1].split(' '))
```

```python
# check if output has already been computed
output_dir = Path('results') / add_underscores_in_spaces(ambiguous_sent)
if os.path.isdir(output_dir):
    if np.all([os.path.isfile(output_dir / f'{add_underscores_in_spaces(d)}.jsonl') for d in disambiguations]):
        print('Continuations already computed!')
        continuation_stats = {}
        for d in disambiguations:
            continuation_stats[d] = pd.read_json(output_dir / f'{add_underscores_in_spaces(d)}.jsonl', lines=True).to_dict(orient='records')
```

```python
generated_continuations = {}
num_generations = 1000

for disambiguation in disambiguations:
    conts = []
    for i in tqdm(range(num_generations//100)):
        gpt3_gens = request(
            prompt=disambiguation,
            stop='\n',
            top_p=0.6,
            model=engine,
            return_only_text=True,
            n=100
        )
        conts.extend(gpt3_gens)
    print(disambiguation + random.sample(conts, 1)[0])
    generated_continuations[disambiguation] = conts
```

```python
# get the minimum number of non-empty continuations
num_conts = num_generations # set to max conts

for disambiguation, continuations in generated_continuations.items():
    non_empty_conts = [c for c in continuations if len(c)>1]
    if len(non_empty_conts) < num_conts:
        num_conts = len(non_empty_conts)
    generated_continuations[disambiguation] = non_empty_conts

print(f'Min number of non-empty continuations: {num_conts}')
for disambiguation, continuations in generated_continuations.items():
    generated_continuations[disambiguation] = random.sample(continuations, num_conts)
```

```python
continuation_stats = {}

for disambiguation, continuations in generated_continuations.items():
    # calculate CE given appropriate disambiguation
    print(disambiguation + random.sample(continuations,1)[0])
    ce_list_cond = cross_entropy([disambiguation]*num_conts, continuations, engine=engine)
    assert len(continuations) == len(ce_list_cond)
    
    # calculate CE given ambiguous prompt
    ce_list_ambig = cross_entropy([ambiguous_sent]*num_conts, continuations, engine=engine)
    assert len(continuations) == len(ce_list_ambig)
    
    continuation_stats[disambiguation] = [
        {'continuation': c, 'ce_cond': x, 'ce_ambig': y, 'pmi_ambig': x-y} 
        for c,x,y in zip(continuations, ce_list_cond, ce_list_ambig)
    ]
```

```python
fig, ax = plt.subplots()
cum_stats = {}
for i, (disambiguation, stats) in enumerate(continuation_stats.items()):
    stat_df = pd.DataFrame(stats)
    cum_stats[disambiguation] = {'mean': stat_df.pmi_ambig.mean(), 'std': stat_df.pmi_ambig.std()}
    sns.histplot(data=stat_df, x='pmi_ambig', color=sns.color_palette('muted')[i], label=disambiguation, kde=True, stat='density')

ax.legend()
```

```python
key = 'mean'
matrix = np.zeros([len(disambiguations)]*2)
for i, (d1, stats1) in enumerate(cum_stats.items()):
    for j, (d2, stats2) in enumerate(cum_stats.items()):
        matrix[i,j] = np.abs(stats1[key] - stats2[key])

mat_df = pd.DataFrame(matrix, columns=cum_stats.keys())
mat_df['Disambiguation'] = cum_stats.keys()
mat_df = mat_df.set_index('Disambiguation')
sns.heatmap(mat_df)
plt.xticks(rotation=45, ha='right')
```

```python
key = 'std'
matrix = np.zeros([len(disambiguations)]*2)
for i, (d1, stats1) in enumerate(cum_stats.items()):
    for j, (d2, stats2) in enumerate(cum_stats.items()):
        matrix[i,j] = np.abs(stats1[key] - stats2[key])

mat_df = pd.DataFrame(matrix, columns=cum_stats.keys())
mat_df['Disambiguation'] = cum_stats.keys()
mat_df = mat_df.set_index('Disambiguation')
sns.heatmap(mat_df)
plt.xticks(rotation=45, ha='right')
```
