```python
import pandas as pd
```

```python
ambient_df = pd.read_json('annotation/AmbiEnt/validated_examples.jsonl', lines=True)
```

```python
rows = []
for i, row in ambient_df.iterrows():
    if 'showed' in row['premise'].lower() and 'showed' not in row['hypothesis'].lower():
        rows.append(row.to_dict())
```

```python
rows
```

```python
rows = []
for i, row in ambient_df.iterrows():
    for sentence_key in ['premise', 'hypothesis']:
        if row[f'{sentence_key}_ambiguous']:
            if any(d[sentence_key] == row[sentence_key] for d in row['disambiguations']):
                rows.append(row.to_dict())
```
